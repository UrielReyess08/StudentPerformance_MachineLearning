```python
from pandas.plotting import scatter_matrix

attributes = ["Hours Studied", "Performance Index", "Sample Question Papers Practiced", "Sleep Hours"]

scatter_matrix(df[attributes], figsize=(12, 8))

plt.tight_layout()
plt.show()
```

```python
df['Performance Index'] = df['Performance Index'].astype('int64')
df.info()
```

```python
corr_matrix["Hours Studied"].sort_values(ascending=False)
```

```python
df["effective_study"] = df["Hours Studied"] * df["Sleep Hours"] #Estudio Efectivo
df["prep_index"] = df["Sample Question Papers Practiced"] * df["Hours Studied"] #Índice de preparación
df["potential_score"] = (df["Previous Scores"] + df["prep_index"]) / 2 #Rendimiento potencial
```

```python
numericos_2 = df.select_dtypes(include=['float64','int64'])

corr_matrix = numericos_2.corr()
corr_matrix["Hours Studied"].sort_values(ascending=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False)

df_cat = df[['Extracurricular Activities']]

df_cat_1hot = cat_encoder.fit_transform(df_cat)

df_cat_1hot_df = pd.DataFrame(df_cat_1hot, columns=cat_encoder.categories_[0].tolist())

df_cat_1hot_df
```

```python
df_feats = df.drop(columns='Extracurricular Activities')

df_feats = pd.concat([df_feats, df_cat_1hot_df], axis=1)

df_feats.tail()
```
