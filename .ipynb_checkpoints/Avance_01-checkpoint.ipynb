```python
# OneHotEncoder: Access_to_Resources
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False)

resources_cat = df_student[['Access_to_Resources']]

resources_cat_1hot = cat_encoder.fit_transform(resources_cat)

feature_name_1 = 'Resources'
new_columns_1 = [f'{val}-{feature_name_1}' for val in cat_encoder.categories_[0].tolist()]

resources_cat_1hot_df = pd.DataFrame(resources_cat_1hot, columns=new_columns_1)

resources_cat_1hot_df.head()
```
