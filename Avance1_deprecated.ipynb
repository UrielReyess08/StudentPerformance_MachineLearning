```python
# OneHotEncoder: Texto a númerico - Gender
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False) # Inicializar

df_cat = df[['Gender']]

df_cat_1 = cat_encoder.fit_transform(df_cat)

df_cat_1_df = pd.DataFrame(df_cat_1, columns=cat_encoder.categories_[0].tolist())

df_cat_1_df
```

```python
# Unir al dataframe original

df_feats = df.drop(columns='Gender')

df_feats = pd.concat([df_feats, df_cat_1_df], axis=1)

df_feats.tail()
```

```python
# OneHotEncoder: Texto a númerico - Sleep Duration
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False) # Inicializar

df_sleep = df[['Sleep Duration']]

df_sleep_1 = cat_encoder.fit_transform(df_sleep)

df_sleep_1_df = pd.DataFrame(df_sleep_1, columns=cat_encoder.categories_[0].tolist())

df_sleep_1_df
```

```python
# OneHotEncoder: Texto a númerico - Dietary Habits
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False) # Inicializar

df_dietary = df[['Dietary Habits']]

df_dietary_1 = cat_encoder.fit_transform(df_dietary)

df_dietary_1_df = pd.DataFrame(df_dietary_1, columns=cat_encoder.categories_[0].tolist())

df_dietary_1_df
```

```python
# OneHotEncoder: Texto a númerico - Have you ever had suicidal thoughts ?
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False) # Inicializar

df_suicidal = df[['Have you ever had suicidal thoughts ?']]

df_suicidal_1 = cat_encoder.fit_transform(df_suicidal)

df_suicidal_1_df = pd.DataFrame(df_suicidal_1, columns=cat_encoder.categories_[0].tolist())

df_suicidal_1_df
```

```python

# OneHotEncoder: Texto a númerico - Family History of Mental Illness
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder(sparse_output=False) # Inicializar

df_family = df[['Family History of Mental Illness']]

df_family_1 = cat_encoder.fit_transform(df_family)

# Agregar prefijo a cada categoría
feature_name = 'Family'
new_columns = [f'{val}-{feature_name}' for val in cat_encoder.categories_[0].tolist()]

# Crear DataFrame con nombres más claros
df_family_1_df = pd.DataFrame(df_family_1, columns=new_columns)

df_family_1_df.head()

#df_family_1_df = pd.DataFrame(df_family_1, columns=cat_encoder.categories_[0].tolist())

#df_family_1_df
```

```python
# Unir al dataframe original

df_feats = df.drop(columns=['Sleep Duration', 'Gender', 'Dietary Habits', 'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness'])

df_feats = pd.concat([df_feats, df_sleep_1_df, df_cat_1_df, df_dietary_1_df, df_suicidal_1_df, df_family_1_df], axis=1)

df_feats.tail()
```
